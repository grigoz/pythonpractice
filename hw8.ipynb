```python
import pandas as pd
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
data=pd.read_csv('wine_data.csv')
```

```python
data.head(10)
```

```python
corr = data.corr()
corr
```

```python
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
sns.pairplot(data, vars=['Class', 'Alcohol'])
```

```python
plt.plot(data['Magnesium'],data['Proline'])
```

```python
plt.plot(data['Class'],data['Total phenols'])
```

```python
class_1 = data[data['Class'] == 1]['Alcohol']
class_2 = data[data['Class'] == 2]['Alcohol']

stats.ttest_ind(class_1, class_2) 
```

```python
class_1 = data[data['Class'] == 1]['Color intensity']
class_2 = data[data['Class'] == 2]['Color intensity']

stats.ttest_ind(class_1, class_2) 
```

```python
class_1 = data[data['Class'] == 1]['Malic acid']
class_2 = data[data['Class'] == 2]['Malic acid']

stats.ttest_ind(class_1, class_2) 
```
